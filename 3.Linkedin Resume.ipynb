```python
import re
def normalize_experience(experience):
    if experience is None:  # Check for None
        return np.nan
    if 'Fresh Graduate' in experience:
        return 0
    elif 'year' in experience or 'month' in experience:
        years = re.findall(r'(\d+)\s*year', experience)
        months = re.findall(r'(\d+)\s*month', experience)
        
        total_years = int(years[0]) if years else 0  
        total_months = int(months[0]) if months else 0  
        
        return total_years + (total_months / 12) 
    
    elif 'year' in experience:
        numbers = re.findall(r'\d+', experience)
        if len(numbers) == 1:
            return int(numbers[0])
        elif len(numbers) == 2:
            return (int(numbers[0]) + int(numbers[1])) / 2 
            
```

```python
data['Normalized Experience'] = data['time_in_position'].apply(normalize_experience)
```

```python
import numpy as np


def get_highest_degree(education_1=None, education_2=None, education_3=None):
    # Helper function to identify the degree level from a single education entry
    def identify_degree(education):
        if pd.isna(education) or education.strip() == '':  # Handle NaN or blank values
            return np.nan
        education = education.lower()  # Convert to lowercase for case-insensitive matching
        if 'master' in education or 'magister' in education:  # English and Indonesian terms for Master's
            return 'Master'
        elif 'bachelor' in education or 'sarjana' in education:  # English and Indonesian terms for Bachelor's
            return 'Bachelor'
        return np.nan  # Return NaN if no degree match is found

    # Identify degrees in all education columns (handle None or empty cases)
    degree_1 = identify_degree(education_1)
    degree_2 = identify_degree(education_2)
    degree_3 = identify_degree(education_3)
    
    # Determine the highest degree: Master's > Bachelor's
    if degree_1 == 'Master' or degree_2 == 'Master' or degree_3 == 'Master':
        return 'Master'
    elif degree_1 == 'Bachelor' or degree_2 == 'Bachelor' or degree_3 == 'Bachelor':
        return 'Bachelor'
    
    return np.nan  # Return NaN if no degrees are found
```

```python
data['highest_degree'] = data.apply(lambda row: get_highest_degree(row['education_1'], 
                                                               row['education_2'], 
                                                               row.get('education_3')), axis=1)
```

```python
from datetime import datetime

def calculate_duration(education_date_1=None, education_date_2=None, education_date_3=None):
    def duration_in_years(date_range):
        if pd.isna(date_range) or date_range.strip() == '':
            return np.nan
        try:
            # Split the date range into start and end dates
            start_date_str, end_date_str = date_range.split(' - ')
            start_date = datetime.strptime(start_date_str.strip(), '%Y-%m-%d')
            end_date = datetime.strptime(end_date_str.strip(), '%Y-%m-%d')
            
            # Calculate the difference in years
            duration = (end_date - start_date).days / 365.25  # Account for leap years
            return round(duration, 2)  # Round to 2 decimal places
        except Exception as e:
            return np.nan  # Return NaN if there's an error with the date format
    
    # Calculate the duration for each education date
    duration_1 = duration_in_years(education_date_1)
    duration_2 = duration_in_years(education_date_2)
    duration_3 = duration_in_years(education_date_3)
    
    return duration_1, duration_2, duration_3
```

```python
data[['duration_1', 'duration_2', 'duration_3']] = data.apply(
    lambda row: pd.Series(calculate_duration(row['education_date_1'], 
                                             row['education_date_2'], 
                                             row['education_date_3'])), axis=1)
```
